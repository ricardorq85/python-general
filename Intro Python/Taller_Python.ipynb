- Cuente cuantas veces se repite cada caracter en el nombre:

Output:

{'d': 1, 'a': 3, 'e': 3, 'n': 2, 'r': 3, 'y': 2, 's': 1, ' ': 1, 't': 1, 'g': 1}

````python
nombre = 'Ricardo Rojas'
nombreDict = {}
for elem in nombre:
    if (elem in nombreDict):
        nombreDict[elem] = nombreDict[elem]+1
    else:
        nombreDict[elem] = 1
    
nombreDict
````

````python
def contarCaracteres (nombre):
    nombreDict = {}
    for elem in nombre:
        if (elem in nombreDict):
            nombreDict[elem] = nombreDict[elem]+1
        else:
            nombreDict[elem] = 1

    return nombreDict
````

````python
contarCaracteres('Juan Fernando Gomez')
````

- Ejercicio: Dada una lista ordenada con números enteros, y un valor de entrada x, encuentre el ceiling de x en la lista, es decir, el valor en la lista más pequeño que sea mayor o igual a x.

  * Por ejemplo si la lista es [1, 2, 8, 10, 10, 12, 19]
  * Para  x = 0:    ceil  = 1
  * Para x = 1:     ceil  = 1
  * Para x = 5:     ceil  = 8
  * Para x = 20:   no hay ningún valor dentro del array que sea el ceil de x

````python
def findCeilOnList(lista, x):
    ceilOnList = None
    for i in lista:
        if (i>= x):
            ceilOnList = i
            break
    return ceilOnList
````

````python
sortedList = [1,2,8,10,10,12,19]
print(findCeilOnList(sortedList, 0))
print(findCeilOnList(sortedList, 1))
print(findCeilOnList(sortedList, 5))
print(findCeilOnList(sortedList, 20))
````
